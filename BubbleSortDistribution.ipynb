## **1. Reproduce the Bubble Sort (optimized – stops if in a round does not do swaps) time distribución calculated in ungrid in a Colab notebook using permutations from itertools and histogram from matplotlib and upload the notebook in BubbleSortDistribution,ipynb in the group’s github.**


```
from itertools import permutations
import time
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from matplotlib import figure
import seaborn as sns
```

```
def bubble_sort(lst):
  pasos = 0
  swapped = False
  n = len(lst)
  for i in range (n):
    swapped = False
    for j in range (0, n - i - 1):
      pasos += 1
      if lst[j] > lst[j + 1]:
        lst[j], lst[j + 1] = lst[j + 1], lst[j]
        pasos += 1
        swapped = True
    if (swapped == False):
      break
  return lst, pasos
```

```
def perm(n):
  lst = (i for i in range (0, n))
  p = permutations(lst)
  return p
```

```
j = 9
for i in range(j):
  times = []
  pasos_list = []
  n = i + 1 # number of elements to permutate
  for i in perm(n):
    # print(i, "desordendado")
    # print(bubble_sort(list(i)))
    time0 = time.time()
    sorted, pasos = bubble_sort(list(i))
    # print(sorted)
    pasos_list.append(pasos)
    times.append(time.time() - time0) 

  title = "Bubble Sort time distribución for sorting the permutations of " + str(n) + " elements" 
  plt.title(title)
  plt.xlabel("Time (s)")
  plt.ylabel("Frecuency")
  if n >= 6 : plt.hist(times, bins=80, alpha=1, edgecolor = 'black',  linewidth=1) 
  else : plt.hist(times, alpha=1, edgecolor = 'black',  linewidth=1)
  plt.grid(True)
  plt.figure(figsize=(10000,10000))
  plt.show()
  plt.clf()
  print("\tMinimun number of steps: ", min(pasos_list))
  print("\tMaximum number of steps: ", max(pasos_list))
  print("\tAverage number of steps: ", sum(pasos_list) / len(pasos_list))
  print("\n")
```

### **Analysis of results**

We can see that the optimization of the Bubble Sort algorithm is successful, because the execution time is reasonable compare to the original code provided by the teacher.

When the number of elements is equal to one, the number of steps is obviously zero, since it is already ordered. 

On the one hand, the best case we have $O(n)$ comparisons and $O(1)$ swaps, for that reason the minimum number of steps is very close to n (number of elements to sort). 

On the other hand, the maximum number of steps has a difference of $8$ (or less) between $n^2$, which is relatively little and the reason is because in the worst case we have $O(n^2)$ comparisons and $O(n^2)$ swaps. 

Finally, the average yield is $O(n^2)$ comparisons and $O(n^2)$ swaps, and that's why the average number of steps is very close to $n^2$.
