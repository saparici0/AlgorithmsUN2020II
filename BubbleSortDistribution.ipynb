```
from itertools import permutations
import time
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
from matplotlib import figure
import seaborn as sns
```

```
def bubble_sort(lst):
  swapped = False
  n = len(lst)
  for i in range (n):
    swapped = False
    for j in range (0, n - i - 1):
      if lst[j] > lst[j + 1]:
        lst[j], lst[j + 1] = lst[j + 1], lst[j]
        swapped = True
    if (swapped == False):
      break
  return lst
```

```
def perm(n):
  lst = (i for i in range (0, n))
  p = permutations(lst)
  return p
```

```
j = 9
for i in range(j):
  times = []
  n = i + 1 # number of elements to permutate
  for i in perm(n):
    # print(i, "desordendado")
    # print(bubble_sort(list(i)))
    time0 = time.time()
    sorted = bubble_sort(list(i))
    # print(sorted)
    times.append(time.time() - time0) 

  title = "Bubble Sort time distribución for sorting the permutations of " + str(n) + " elements" 
  plt.title(title)
  plt.xlabel("Time (s)")
  plt.ylabel("Frecuency")
  if n >= 6 : plt.hist(times, bins=80, alpha=1, edgecolor = 'black',  linewidth=1) 
  else : plt.hist(times, alpha=1, edgecolor = 'black',  linewidth=1)
  plt.grid(True)
  plt.figure(figsize=(10000,10000))
  plt.show();
  plt.clf();
```
